```python
import pandas as pd
```

```python
df = pd.read_csv('results.txt', sep='\t')
df.loc[df.grb_threads == 1,"algorithm"] = "1 grb thread"
df.loc[df.grb_threads == 2,"algorithm"] = "2 grb threads"
df.loc[df.grb_threads == 4,"algorithm"] = "4 grb threads"
df.loc[df.grb_threads == 8,"algorithm"] = "8 grb threads"
df.loc[df.grb_threads == 16,"algorithm"] = "16 grb threads"
df["dataset"] = "panacea"
df["size"] = 1
df[["algorithm", "dataset", "seed", "size", "cpus", "time"]].to_csv('data.tsv', sep='\t', index=None)
```
